```python
sns.boxplot(x='artist_pop',data=rd)
```

```python
md = rd.drop(['artist_pop','track_pop'], axis=1)
```

We can use the drop command to form a new dataset by removing a column

```python
corelation = md.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```




A heatmap is used for finding out correlation between different attributes
Here we can see how danceability, energy and tempo are in high cohesion with each other whereas valence mode and key are not on the other hand

```python
sns.pairplot(md)
```

A pairplot pairs up all the attributes in a sequential order and then describes their values with each other as a scatterplot 

```python
sns.relplot(x='energy',y='danceability',hue='key',data=md)
```

Relplot command also gives us scatterplots but with specified attributes and with a common relative grading attribute here we can see how energy and danceability are related in relevance to key

```python
sns.relplot(x='tempo',y='acousticness',hue='key',data=md)
```

Here high key songs are seen to have mid range tempo with a few exceptions but the acousticness is generally seen to be low in hit songs

```python
sns.relplot(x='artist_pop',y='track_pop',hue='key',data=rd)
```

Since the dataset taken is of hit songs the track pop rating is seen to be high and also the same for the artist pop rating as well but there are few bangers from low key artists as well

```python
sns.relplot(x='artist_pop',y='track_pop',hue='mode',data=rd)
```

Here we see the same graph but with a different relative parameter mode so the spread is the same but we can infer the specific mode and the trend that it follows

```python
sns.relplot(x='energy',y='valence',hue='mode',data=rd)
```

```python
sns.distplot(md['energy'])
```

Distplot shows us the distribution in a particular attribute in comparison to other attributes. Here we see the example of energy when compared against density of the data set. There is a peak in value at 0.7 energy and the graph is tapering on both sides of the peak

```python
sns.distplot(md['tempo'])
```

We see a lot of songs gathered around the tempo of 125 but there are also a few hits at 175 tempo rating

```python
sns.catplot(x='danceability', kind='box', data=rd)
```

Catplot is a plot that has the range of values ranging from the 25th and 75th percentile value

```python
sns.catplot(x='track_pop', kind='box', data=rd)
```

```python
plt.figure(figsize=(20,40))
barplotter = sns.barplot(data=rd, x="artist_pop", y="artist_name");
```
